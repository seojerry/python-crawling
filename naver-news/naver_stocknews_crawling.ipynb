```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time
import re
import requests
chrome_options = webdriver.ChromeOptions()

#크롬 Webdriver객체 생성
wd = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
news_url_list=[]
news_title_list=[]
news_contents_list=[]
news_dates_list =[]
#ConnectionError방지
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) Chrome/98.0.4758.102"}
for page in range(1,5):
    naver_url='https://news.naver.com/main/list.naver?mode=LS2D&sid2=258&sid1=101&mid=shm&date=20230208&page=%d'%(page)
    wd.get(naver_url)
    time.sleep(1)
    html = wd.page_source
    soup = BeautifulSoup(html,'html.parser')
    news_urls = soup.select('ul.type06_headline > li > dl > dt > a')
    for i in news_urls:
        news_url_list.append(i.attrs['href'])

news_url_list = list(set(news_url_list))

news_list = list()

for i in news_url_list:
    news_dict = dict()
    news = requests.get(i,headers=headers)
    news_html = BeautifulSoup(news.text,'html.parser')

    #뉴스 제목 가져오기
    title = news_html.select_one('#title_area > span')
    content = news_html.select_one('#dic_area')
    date = news_html.select_one('div.media_end_head_info_datestamp_bunch > span')
    link = i
    title = re.sub('<.*?>','',str(title)).strip()
    content = re.sub('<.*?>','',str(content)).strip()
    date = re.sub('<.*?>','',str(date))
    news_dict["title"] = title
    news_dict["content"] = content
    news_dict["date"] = date
    news_dict["link"] = link
    news_list.append(news_dict)

print(news_list)
    
    
        
```
